```python
import pandas as pd
import re
import os

folder_path = "../kursach/alldata"


combined_data = ""


for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, 'r') as file:
            file_content = file.read()

        cleaned_data = file_content.replace("P", "")
        cleaned_data = re.sub(r'(\s{8})+', '?', cleaned_data)
        cleaned_data = re.sub(r'(\s{6})+', 'kal?', cleaned_data)
        
        combined_data += cleaned_data


# Зберегти об'єднані дані у файл

output_file_path = '../kursach/alldata.txt'
with open(output_file_path, 'w') as output_file:
    output_file.write(combined_data)
```

```python
pattern = r"-[0-9]+-[0-9]+"

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            file_content = file.read()

            if re.search(pattern, file_content):
                print(f"Файл '{filename}' містить рядок у форматі '-number-number'")
                break 
else:
    print("Файл з рядком у форматі '-number-number' не знайдено в папці")
```

```python
df[["HA_Start", "HA_End"]] = df['H_Angle'].str.split('-', expand=True)
df[["VA_Start", "VA_End"]] = df['V_Angle'].str.split('-', expand=True)
df = df.drop("V_Angle",  axis=1)
df = df.drop("H_Angle",  axis=1)
df
```

Варік з обрізанням данних по вологості


```python
df = df.drop(df[df['Humid'].str.contains('kal')].index)
df = df.sample(frac = 1)
df['Temp'] = df['Temp'].str.replace(',', '.').astype(float)
df['Temp'] = pd.to_numeric(df['Temp'])
df[["Temp", "Humid", "HA_Start", "HA_End","VA_Start", "VA_End"]] = df[["Temp", "Humid", "HA_Start", "HA_End","VA_Start", "VA_End"]].apply(pd.to_numeric)
df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures


train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
model = LinearRegression(fit_intercept=True)
poly_features = PolynomialFeatures(degree=4)
x_train = train_df[['Range', 'Temp', 'Humid', 'Hight']]
y_train = train_df[['HA_Start']]

x_poly = poly_features.fit_transform(x_train)
model.fit(x_poly, y_train)
predicted = model.predict(x_poly)
print('На тренувальних даних MSE =', mean_squared_error(y_train, predicted))
```
