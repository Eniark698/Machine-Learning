```python
import pandas as pd
import re
import os

folder_path = "../kursach/alldata"


combined_data = ""


for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, 'r') as file:
            file_content = file.read()

        cleaned_data = file_content.replace("P", "")
        cleaned_data = re.sub(r'(\s{8})+', '?', cleaned_data)
        cleaned_data = re.sub(r'(\s{6})+', 'NaN?', cleaned_data)
        
        combined_data += cleaned_data



output_file_path = '../kursach/alldata.txt'
with open(output_file_path, 'w') as output_file:
    output_file.write(combined_data)
```

```python
folder_path = "../kursach/alldata"
df = pd.read_csv('../kursach/alldata.txt', names=["Time","H_Angle","V_Angle","Range","Temp","Humid","Hight"], sep="?")
df
```

```python

df = df.drop("Time", axis=1)

df.info()
with open("../kursach/test_help.txt", 'w') as file:
    file.write("")
df.to_csv("../kursach/test_help.txt", index=False)
df
```

```python
pattern = r"-[0-9]+-[0-9]+"

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            file_content = file.read()

            if re.search(pattern, file_content):
                print(f"Файл '{filename}' містить рядок у форматі '-number-number'")
                break 
else:
    print("Файл з рядком у форматі '-number-number' не знайдено в папці")
```

```python
df[["HA_Start", "HA_End"]] = df['H_Angle'].str.split('-', expand=True)
df[["VA_Start", "VA_End"]] = df['V_Angle'].str.split('-', expand=True)
df = df.drop("V_Angle",  axis=1)
df = df.drop("H_Angle",  axis=1)

df
```

```python
df['Temp'] = df['Temp'].str.replace(',', '.').astype(float)
df['Temp'] = pd.to_numeric(df['Temp'])
df[["Temp",  "HA_Start", "HA_End","VA_Start", "VA_End"]] = df[["Temp", "HA_Start", "HA_End","VA_Start", "VA_End"]].apply(pd.to_numeric)
df.info()
df.to_csv('output_file.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
model = Lasso(alpha=0.1)
poly_features = PolynomialFeatures(degree=3)

x_train = train_df[['Range', 'Temp', 'Hight']]
y_train = train_df[['HA_Start', 'HA_End', 'VA_Start', 'VA_End']]

x_test= test_df[['Range', 'Temp', 'Hight']]
y_test = test_df[['HA_Start', 'HA_End', 'VA_Start', 'VA_End']]

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

x_train_poly = poly_features.fit_transform(x_train)
model.fit(x_train_poly , y_train)
train_predicted = model.predict(x_train_poly)

x_test_poly = poly_features.fit_transform(x_test)
model.fit(x_test_poly , y_test)
test_predicted = model.predict(x_test_poly)


print('На тренувальних даних MSE =', mean_squared_error(y_train, train_predicted))
print('На тестових даних MSE =', mean_squared_error(y_test, test_predicted))
predicted_example = model.predict(poly_features.fit_transform([[11570, -8.7, 4621]]))
print("Приклад передбаченних даних: ", predicted_example[0][0], predicted_example[0][1], predicted_example[0][2], predicted_example[0][3])
```
