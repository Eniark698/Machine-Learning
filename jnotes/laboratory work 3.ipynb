```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
df = pd.read_csv('../datasets/Social_Network_Ads.csv')
df = df.sample(frac=1)
# Calculate the percentage of null values in the specified column
```

```python
xData, yData  = df[["Age", "EstimatedSalary"]], df["Purchased"] 
```

```python

l = len(xData)
test_x = np.array(xData.iloc[:round(0.2*l)])
train_x = np.array(xData.iloc[round(0.2*l):])

l = len(yData)
test_y = np.array(yData.iloc[:round(0.2*l)])
train_y = np.array(yData.iloc[round(0.2*l):])
model = LogisticRegression()


plt.plot(train_x[train_y == 1, 0], train_x[train_y == 1, 1], 'k+', label="Прийнятий")
plt.plot(train_x[train_y == 0, 0], train_x[train_y == 0, 1], 'yo', label="Не прийнятий")

model.fit(train_x, train_y)

print('Точність моделі з тестовим сетом: ' + str(model.score(train_x, train_y)*100) + 
'%\n')
print('Чутливість моделі: ' + str(metrics.recall_score(train_y, model.predict(train_x))*100) + '%\n')
print('Специфічність моделі: '+ str(metrics.precision_score(train_y, model.predict(train_x))*100) + '%')
```

```python
print('Точність моделі з тестовим сетом: ' + str(model.score(test_x, test_y)*100) + 
'%\n')
print('Чутливість моделі: ' + str(metrics.recall_score(test_y, model.predict(test_x))))
print('Специфічність моделі: '+ str(metrics.precision_score(test_y, model.predict(test_x))*100) + '%')
```
